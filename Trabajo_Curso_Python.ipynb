```python
import fitz #PyMuPDF
import pandas as pd
import sqlite3

documento = fitz.open('/content/Categorias.pdf')

texto_completo = ""
for pagina_num in range(len(documento)):
  pagina = documento[pagina_num]
  texto_completo += pagina.get_text()
  print(texto_completo)

#Para la data sin estructurar
categorias_data = []

# Para extraer linea por linea
for line in texto_completo.split('\n'):
    line = line.strip() # Remover espacios
    if line and not line.startswith('ID Categoria'): # Saltar lineas vacias
        parts = line.split(maxsplit=1)
        if len(parts) == 2:
            try:
                categoria_id = int(parts[0])
                categoria_name = parts[1]
                categorias_data.append({'ID': categoria_id, 'Categorias': categoria_name})
            except ValueError:
                # Handle cases where the first part isn't a valid ID
                continue

# Crear el DF de la información ordenada, asignándolo a df_categorias
df_categorias = pd.DataFrame(categorias_data)

display(df_categorias)

display(df_excel)
```

```python
# Asegurarse de que la columna 'Categoría' en df_excel sea del mismo tipo que 'ID' en df_categorias
df_excel['Categoría'] = df_excel['Categoría'].astype(int)

# Unir df_excel con df_categorias para obtener los nombres de las categorías
df_excel = pd.merge(
    df_excel,
    df_categorias[['ID', 'Categorias']],
    left_on='Categoría',
    right_on='ID',
    how='left'
)

# Reemplazar la columna 'Categoría' original con los nombres de las categorías
df_excel['Categoría'] = df_excel['Categorias']

# Eliminar las columnas auxiliares creadas por el merge ('ID' y 'Categorias_y' si se crea por el merge)
df_excel = df_excel.drop(columns=['ID', 'Categorias_y'], errors='ignore')

# Renombrar 'Categorias_x' a 'Categorias' (si es que se creó por el merge)
df_excel = df_excel.rename(columns={'Categorias_x': 'Categorias'}, errors='ignore')

df_excel_categorias = df_excel.drop(columns=['Categorias'])
display(df_excel_categorias)
```

```python
#Cálculo de métricas

# Calcular el precio total de venta por cada transacción (fila)
df_excel_categorias['Total Venta por Transacción'] = df_excel_categorias['Cantidad'] * df_excel_categorias['Precio Venta']

# Calcular el total de ventas global sumando todas las ventas por transacción
total_ventas_global = df_excel_categorias['Total Venta por Transacción'].sum()

df_excel_categorias['Total Costo por Transacción'] = df_excel_categorias['Cantidad'] * df_excel_categorias['Costo Venta']
total_costo_global = df_excel_categorias['Total Costo por Transacción'].sum()

Margen = (total_ventas_global - total_costo_global)

print(f"Total de Ventas Global: {total_ventas_global:.2f}")
print(f"Total de Costo Global: {total_costo_global:.2f}")
print(f"Margen Total: {Margen:.2f}")

display(df_excel_categorias)
```

```python
#contrucción de modelo dimensional (esquema de estrella)

#generar dataframe para cada columna de df_excel_categorias

#asegurarse que la columna 'Fecha' en df_excel_categorias sea tipo datatime

df_excel_categorias['Fecha'] = pd.to_datetime(df_excel_categorias['Fecha'])

df_Fecha_dim = df_excel_categorias['Fecha'].drop_duplicates().reset_index(drop=True)

df_Factura_dim = df_excel_categorias['Factura'].drop_duplicates().reset_index(drop=True)

df_SKU_dim = df_excel_categorias['SKU'].drop_duplicates().reset_index(drop=True)

df_Producto_completo_dim = df_excel_categorias['Producto Completo'].drop_duplicates().reset_index(drop=True)

df_Marca_dim = df_excel_categorias['Marca'].drop_duplicates().reset_index(drop=True)

df_Categoria_Final_dim = df_excel_categorias['Categoría'].drop_duplicates().reset_index(drop=True)

df_Tipo_de_compra_dim= df_excel_categorias['Tipo de Compra'].drop_duplicates().reset_index(drop=True)

df_Cantidad_dim = df_excel_categorias['Cantidad'].drop_duplicates().reset_index(drop=True)

df_Precio_venta_dim = df_excel_categorias['Precio Venta'].drop_duplicates().reset_index(drop=True)

df_Costo_venta_dim= df_excel_categorias['Costo Venta'].drop_duplicates().reset_index(drop=True)

display(df_Fecha_dim)
```

```python
import pandas as pd
import sqlite3

#conectar a la base de datos
conn_model = sqlite3.connect('modelo_ventas.db')
print("Conexión a la base de datos 'modelo_ventas.db' establecida.")

#consulta SQL para calcular el margen total por categoría en la bd
query = """
SELECT
    dc.Categoría AS Nombre_Categoria,
    SUM(ff."Total Venta por Transacción") AS Total_Ventas,
    SUM(ff."Total Costo por Transacción") AS Total_Costos,
    SUM(ff."Total Venta por Transacción") - SUM(ff."Total Costo por Transacción") AS Margen_Total
FROM
    fact_table_final AS ff
JOIN
    dim_categoria_final AS dc ON ff.Categoria_Final_ID = dc.Categoria_Final_ID
GROUP BY
    dc.Categoría
ORDER BY
    Margen_Total DESC;
"""

# Ejecutar la consulta y cargar los resultados en un DataFrame de pandas
margen_por_categoria = pd.read_sql_query(query, conn_model)

# Cerrar la conexión a la base de datos
conn_model.close()

# Mostrar el resultado
display(margen_por_categoria)
```

```python
import sqlite3
import pandas as pd

# Conectar a la base de datos
conn_model = sqlite3.connect('modelo_ventas.db')
print("Conexión a la base de datos 'modelo_ventas.db' establecida.")

# Consulta SQL para calcular el total de ventas por tipo de compra
query_ventas_por_tipo = """
SELECT
    dtc."Tipo de Compra" AS Tipo_de_Compra,
    SUM(ff."Total Venta por Transacción") AS Total_Ventas
FROM
    fact_table_final AS ff
JOIN
    dim_tipo_compra AS dtc ON ff.Tipo_de_Compra_ID = dtc.Tipo_de_Compra_ID
GROUP BY
    dtc."Tipo de Compra"
ORDER BY
    Total_Ventas DESC;
"""

# Ejecutar la consulta y cargar los resultados en un DataFrame de pandas
ventas_por_tipo_compra = pd.read_sql_query(query_ventas_por_tipo, conn_model)

# Cerrar la conexión a la base de datos
conn_model.close()

# Mostrar el resultado
display(ventas_por_tipo_compra)
```
