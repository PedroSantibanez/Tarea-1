```python
# Recopilación de información personal
nombre = input ("ingresa tu nombre")
edad = input ("ingresa tu edad")
color_favorito = input ("ingresa tu color favorito")
lista_de_pasatiempos = input ("ingresa tus pasatiempos (separados por comas):")
```

```python
# Calcular edad aproximada en "años perro"
edad_años_perro = int(edad) * 7
mensaje = f"Mi edad en años perro es: {edad_años_perro}"
print (mensaje)
```

```python
# Calcular longitud del nombre
longitud_nombre = len(nombre)
mensaje = f"La longitud del nombre es: {longitud_nombre} caracteres."
print(mensaje)
```

```python
# extraer primer y último carácter del color favorito
primer_caracter = color_favorito[0]
ultimo_caracter = color_favorito[3]
mensaje = f"El primer caracter del color favorito es {primer_caracter} y el último caracter es {ultimo_caracter}"
print(mensaje)
```

```python
# Convierte la cadena de pasatiempos en una lista de pasatiempos individuales
pasatiempos = lista_de_pasatiempos.split(",")
mensaje = f"Tu lista de pasatiempos es: {pasatiempos}"
print (mensaje)
```
